Fib$(n)$ is the closest integer to
$$\frac{\Phi^n}{\sqrt{5}}$$
where
$$\Phi = \frac{1 + \sqrt{5}}{2}$$
Also,
$$Fib(n) = \frac{\Phi^n - \Upsilon^n}{\sqrt{5}}$$
where
$$\Upsilon = \frac{1 - \sqrt{5}}{2}$$

```python
from matplotlib import pyplot as plt 
%matplotlib inline
from math import *

phi = (1 + sqrt(5)) / 2
def phifunc(n):
    return (phi ** n)/sqrt(5)

upsilon = (1 - sqrt(5)) / 2
```

```python
# List for fibonacci numbers
fib = [0, 1]

# List for Phi^n / sqrt(5)
closeto = [0, phifunc(1)]

for i in range(2, 50):
    fib.append(fib[i - 1] + fib[i - 2])
    closeto.append(phifunc(i))
```

```python
plt.rcParams["figure.figsize"] = (8, 6)
plt.rcParams["figure.dpi"] = 300
plt.scatter(range(1, 10), fib[1:10], label="fib(n)")
plt.scatter(range(1,10), closeto[1:10], label="phifunc(n)")
plt.xlabel('iteration n')
plt.ylabel('result')
plt.legend()
```

Printed differences between Fib$(n)$ and the $\Phi$ equation

```python
for i, (f, c) in enumerate(zip(fib[1:10], closeto[1:10])):
    print(abs(f) - abs(c))
```
