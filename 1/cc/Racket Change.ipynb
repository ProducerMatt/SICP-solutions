```racket
;; Output 1000 iterations with different prime denominations
;; cursed globals
(define fcalls 0)
;; Returns a list with:
;; 1: the amount of change
;; 2: the number of ways change can be made
;; 3: the number of recursions
;; "Helper" for (cc)
(define (count-change amount)
  (set! fcalls 0)
  (let ((ways (cc amount 5)))
    (list amount ways fcalls)))

;; Split recursive definition. Can you write an iterative version?
(define (cc amount kinds-of-coins)
  (set! fcalls (+ fcalls 1))
  (cond ((= amount 0) 1)
        ((or (< amount 0)
             (= kinds-of-coins 0))
         0)
        (else
         (+ (cc amount (- kinds-of-coins 1))
            (cc (- amount (first-denomination
                           kinds-of-coins))
                kinds-of-coins)))))

(define (first-denomination kinds-of-coins)
  (cond ((= kinds-of-coins 1) 5)
        ((= kinds-of-coins 2) 11)
        ((= kinds-of-coins 3) 23)
        ((= kinds-of-coins 4) 113)
        ((= kinds-of-coins 5) 1)))
```

```racket
(define results '())
(define beginning 1)
(define ending 101)
(define (countit) (iota (- ending beginning) beginning 1))

(set! results (map count-change (countit)))

(list results)
```

```racket
(plot-pict (function (lambda x (car (cddr (apply count-change x)))) 1 100 #:label "y = x^2"))
```
